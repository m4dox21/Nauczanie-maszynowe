```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

# Inicjalizacja i nauka modelu KNN
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)

# Predykcje dla zbioru testowego
y_pred_knn = knn_model.predict(X_test)

# Ocena dokÅ‚adnoÅ›ci dla modelu KNN
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"Accuracy for KNN: {accuracy_knn}")
print(classification_report(y_test, y_pred_knn))

# Confusion matrix dla KNN
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)

# ConfusionMatrixDisplay dla KNN
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_knn, display_labels=knn_model.classes_)
disp.plot()
plt.title('Confusion Matrix for KNN')
plt.show()

# Cross-validation scores dla KNN
cv_scores_knn = cross_val_score(knn_model, X_train, y_train, cv=5)
print("Cross-validation Scores for KNN:", cv_scores_knn)
```
